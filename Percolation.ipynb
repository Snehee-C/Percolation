```python
import numpy as np
import random
import time 

n = int(input('Enter the grid size: '))
grid = np.zeros((n,n),dtype = int)

def rand():
    global r
    r = random.randint(1,n**2)
    b = r//n
    c = r%n
    if c==0:
        if grid[b-1][n-1] == 0:
            grid[b-1][n-1] = r
            isOpen(b-1,n-1)
        else:
            rand()
    else:
        if grid[b][c-1] == 0:
            grid[b][c-1] = r
            isOpen(b,c-1)
        else:
            rand()
    
def isOpen(p,q):
    if p==0 and q==0:
        grid[p][q] = -1
        if grid[p][q+1] != 0:
            a=grid[p][q+1] = -1
            Small(grid[p][q],a)
        if grid[p+1][q] != 0:
            a=grid[p+1][q] = -1
            Small(grid[p][q],a)
        
    elif p==0 and q==(n-1):
        grid[p][q] = -1
        if grid[p][q-1] != 0:
            a=grid[p][q-1] = -1
            Small(grid[p][q],a)
        if grid[p+1][q] != 0:
            a=grid[p+1][q] = -1
            Small(grid[p][q],a)
    
    elif p==(n-1) and q==0:
        if grid[p-1][q] != 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
        if grid[p][q+1] != 0:
            a=grid[p][q+1]
            Small(grid[p][q],a)
        
    elif p==(n-1) and q==(n-1):
        if grid[p][q-1] != 0:
            a=grid[p][q-1]
            Small(grid[p][q],a)
        if grid[p-1][q] != 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
    
    elif p==0 and q!=0 and q!=(n-1):
        grid[p][q] = -1
        if grid[p][q-1] != 0:
            a=grid[p][q-1] = -1
            Small(grid[p][q],a)
        if grid[p][q+1] != 0:
            a=grid[p][q+1] = -1
            Small(grid[p][q],a)
        if grid[p+1][q] != 0:
            a=grid[p+1][q] = -1
            Small(grid[p][q],a)
        
    elif p==(n-1) and q!=0 and q!=(n-1):
        if grid[p][q-1] != 0:
            a=grid[p][q-1]
            Small(grid[p][q],a)
        if grid[p][q+1] != 0:
            a=grid[p][q+1]
            Small(grid[p][q],a)
        if grid[p-1][q] != 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
            
    elif q==0 and p!=0 and p!=(n-1):
        if grid[p+1][q] != 0:
            a=grid[p+1][q]
            Small(grid[p][q],a)
        if grid[p][q+1] != 0:
            a=grid[p][q+1]
            Small(grid[p][q],a)
        if grid[p-1][q] != 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
            
    elif q==(n-1) and p!=0 and p!=(n-1):
        if grid[p-1][q] != 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
        if grid[p][q-1]!= 0:
            a=grid[p][q-1]
            Small(grid[p][q],a)
        if grid[p+1][q]!= 0:
            a=grid[p+1][q]
            Small(grid[p][q],a)
            
    else:
        if grid[p+1][q]!= 0:
            a=grid[p+1][q]
            Small(grid[p][q],a)
        if grid[p-1][q]!= 0:
            a=grid[p-1][q]
            Small(grid[p][q],a)
        if grid[p][q+1]!= 0:
            a=grid[p][q+1]
            Small(grid[p][q],a)
        if grid[p][q-1]!= 0:
            a=grid[p][q-1]
            Small(grid[p][q],a)

def Small(x,y):
    s = 0
    if x>y:
        s = y
    else:
        s = x
    for i in range(0,n):
        for j in range(0,n):
            if grid[i][j]==x: 
                grid[i][j]=s
            elif grid[i][j]==y:
                grid[i][j]=s
            
def Check():
    for i in range(0,n-1):
        if grid[n-1][i] == -1:
            return True
        else:
            return False

count = 0
start = time.time()
while Check() == False:
    count = count +1
    rand()
    #print(grid)
print(grid)
print("Time taken: ",time.time()-start)
print("No. of blocks opened: ",count)
print("Percentage:",(count/(n**2))*100)
```
